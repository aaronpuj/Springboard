```python
import pandas as pd
data= pd.read_csv(r'C:\Users\Pujanande\Desktop\Capstone 1\cap1_training_data.csv') #read in training data
```

```python
data['price_round'] = round(data['PRICEINDOLPERMWH']) #round price to nearest dollar
data['MW_round'] = round(data['QUANTITYINMWH']) #round quantity to nearest dollar

data['TRADE_DATE'] = pd.to_datetime(data['TRADE_DATE']) #convert trade date to datetime

data['month'] = data['TRADE_DATE'].dt.month #extract month
data['day'] = data['TRADE_DATE'].dt.weekday #extract day

data['HOUR_FREQ_cat'] = data['HOUR_FREQ'].apply(str) #create str so that integer variables will be picked up as catagorical
data['HOUROFDAY_cat'] = data['HOUROFDAY'].apply(str)
data['MW_round_cat'] = data['MW_round'].apply(str)
data['price_round_cat'] = data['price_round'].apply(str)
```

```python
final = pd.get_dummies(data) #get dummies for all catagorical data
final.head()
```
