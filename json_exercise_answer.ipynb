## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
## question 1

#import data as string
import json
from pandas.io.json import json_normalize
json_data = json.load((open('data/world_bank_projects.json')))

#normalize by country & check
data_country_norm = json_normalize(json_data, '_id', ['countryname'])
data_country_norm
```

```python
#convert to pandas data frame
import pandas as pd
country_df = pd.DataFrame(data_country_norm)

#use group by to sum country totals and display the top ten
country_df_sort = country_df.groupby('countryname').size().sort_values(ascending=False)
country_df_sort.head(10)
```

```python
#remove Africa from inital data frame since it's not a country
answer1_df = country_df[country_df.countryname != "Africa"]

#use group by to sum country totals and display the top ten
answer1_df = answer1_df.groupby('countryname').size().sort_values(ascending=False)
answer1_df.head(10)
```

```python
## question 2

#import data as string
json_data2 = json.load((open('data/world_bank_projects.json')))

#normalize by mjtheme_namecode & check
data_theme_normal = json_normalize(json_data2, 'mjtheme_namecode')
data_theme_normal
```

```python
#convert to pandas data frame
data_theme_df = pd.DataFrame(data_theme_normal)

#use group by to sum country totals and display the top ten
data_theme_sort = data_theme_df.groupby('name').size().sort_values(ascending=False)
data_theme_sort
```

```python
#remove blanks from inital data frame
answer2_df = data_theme_df[data_theme_df.name > " "]

#use group by to sum theme totals and display the top ten
answer2_df = answer2_df.groupby('name').size().sort_values(ascending=False)
answer2_df
```

```python
## question 3

#import data as string 
json_data3 = json.load((open('data/world_bank_projects.json')))

#normalize by mjtheme_namecode & check
data_theme_normal3 = json_normalize(json_data3, 'mjtheme_namecode')
data_theme_normal3
```

```python
#convert to pandas data frame
theme_df = pd.DataFrame(data_theme_normal3)

#drop duplicates and remove blanks to make dataframe with unique paired values and check
theme_df_dup = theme_df.drop_duplicates()
theme_df_small = theme_df_dup[theme_df_dup.name > " "]
theme_df_small
```

```python
#use paired vallue data frame to build dictionary and check
codes_list = theme_df_small['code'].tolist()
names_list = theme_df_small['name'].tolist()
legend = dict(zip(codes_list, names_list))
legend
```

```python
#convert code column to series, use legend dictionary to map correct values to each code and check
theme_series = theme_df['code'].map(legend)
theme_series
```

```python
#convert series back to df and use group by to sum theme totals and display the top ten
answer3_df =pd.DataFrame(theme_series)
answer3_df = answer3_df.groupby('code').size().sort_values(ascending=False)
answer3_df
```
